```python
mixcr_agg_sh = pd.read_csv('../../data_tables/mixcr/shipp_samples_mixcr_aggregated.tsv', sep='\t', index_col=0)
mixcr_agg_st = pd.read_csv('../../data_tables/mixcr/staudt_samples_mixcr_aggregated.tsv', sep='\t', index_col=0)

shipp_s = list(set(mixcr_agg_sh.index))
staudt_s = list(set(mixcr_agg_st.index))

mixcr_agg = pd.concat([mixcr_agg_sh, mixcr_agg_st])
mixcr_agg
```

```python
plt.hist(mixcr_agg.loc[mixcr_agg.index.isin(shipp_s)].index.value_counts(), alpha=0.4, label='shipp', bins=50)
plt.hist(mixcr_agg.loc[mixcr_agg.index.isin(staudt_s)].index.value_counts(), alpha=0.4, label='staudt', bins=50)
plt.title('Unique Clones per Sample')
plt.ylabel('Sample Count')
plt.xlabel('Unique Clones')
plt.legend()
plt.show()
```

```python
plt.hist(mixcr_agg.loc[mixcr_agg.index.isin(shipp_s), 'cloneCount'], 
         alpha=0.4, label='shipp', bins=50)
plt.hist(mixcr_agg.loc[mixcr_agg.index.isin(staudt_s), 'cloneCount'], 
         alpha=0.4, label='staudt', bins=50)
plt.title('Clone Count per Specific Clone')
plt.yscale('log')
#plt.ylabel('Amount')
plt.xlabel('Clone Count')
plt.legend()
plt.show()
```
