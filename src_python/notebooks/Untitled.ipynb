```python
import pandas as pd
```

```python
panel_sets = pd.read_csv('../../data_tables/sample_sets/ShippStaudtSets.purity0.2.txt', sep='\t', index_col=0)
panel_sets
```

```python
gsm = pd.read_csv('../../data_tables/gsm/DLBCL.699.163drivers.Sep_23_2022.tsv', sep='\t', index_col=1)

gsm = gsm.drop('gene', axis=1)
gsm
```

```python
staudt_s = panel_sets.loc[(panel_sets['cohort'] == 'Staudt') | (panel_sets['cohort'] == 'TCGA')].index
gsm_staudt = gsm.loc[:, gsm.columns.isin(staudt_s)]
gsm_shipp = gsm.loc[:, ~gsm.columns.isin(staudt_s)]
gsm_staudt.shape, gsm_shipp.shape
```

```python
gsm_staudt.to_csv('../../data_tables/gsm/DLBCL.422.163drivers.Schmitz.tsv', sep='\t')
gsm_shipp.to_csv('../../data_tables/gsm/DLBCL.277.163drivers.Shipp.tsv', sep='\t')
```
