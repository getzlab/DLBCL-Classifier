```python
our_gsm = pd.read_csv('../../data_tables/gsm/DLBCL.699.163drivers.Sep_23_2022.tsv', sep='\t', index_col=1)
our_gsm_genes = list(our_gsm['gene'])
our_gsm_genes
```

```python
gene_overlap = set(polarix_set).intersection(our_gsm_genes)
gene_overlap, len(gene_overlap)
```

```python
test_set = list(pd.read_csv('../../data_tables/train_test_sets/TestingSet_149Subset_May2021.txt', 
                       sep='\t', index_col=0, header=None).index)
len(test_set)
```

```python
full_gsm_polarix = our_gsm[test_set]
full_gsm_polarix
```

```python
full_gsm_polarix.loc[~full_gsm_polarix.index.isin(gene_overlap)] = 0
full_gsm_polarix.sum(axis=1).sort_values(ascending=False)[0:25]
```

```python
labels = pd.read_csv('../data_tables/confidence_tables/baseline_probabilities.connectivity_based.sensitivity_power2.Sep_23_2022.tsv',
                    sep='\t', index_col=0)
labels['cluster'] = 'C' + labels['cluster'].astype(str)
labels = labels.loc[full_gsm_polarix.columns]
labels
```

```python
# Make reduced gsms

reduced_gsm_polarix = fd.construct_reduced_winning_version(full_gsm_polarix)
classified_samples_polarix = cg.classify_samples_winning_model(reduced_gsm_polarix)
classified_samples_polarix.index.name = 'sample'
classified_samples_polarix['PredictedCluster'] = classified_samples_polarix['PredictedCluster'].map({1.0: 'C1', 
                                                                                                     2.0: 'C2', 
                                                                                                     3.0: 'C3', 
                                                                                                     4.0: 'C4', 
                                                                                                     5.0: 'C5'})
classified_samples_polarix['TrueCluster'] = labels['cluster']
classified_samples_polarix = classified_samples_polarix.sort_values(by='Confidence')
classified_samples_polarix.to_csv('../evaluation_panel_sets/phoenix_predictions.tsv', sep='\t', index=True)
```

```python
performances_all = pd.DataFrame(columns=['accuracy', 'lowerAcc', 'upperAcc', 
                                         'kappa', 'lowerKappa', 'upperKappa',
                                         'performance', 'lowerPerformance', 'upperPerformance'])
```

```python
# f1 cdx results

lrx_polarix = np.array(classified_samples_polarix['Confidence'])
lry_polarix = np.array(np.equal(classified_samples_polarix['PredictedCluster'], 
                                classified_samples_polarix['TrueCluster']).astype(int))
CM.modelKappa(lrx_polarix, lry_polarix,
              num_iter=1,
              computeplots=True,
              savefilename='../plots/xyresiduals/phoenix')

acc_polarix, lower_acc_polarix, upper_acc_polarix = CM.modelAccuracy(classified_samples_polarix['PredictedCluster'], 
                                                                     classified_samples_polarix['TrueCluster'])

kappa_polarix, lower_kappa_polarix, upper_kappa_polarix = CM.modelKappa(lrx_polarix, lry_polarix)

p_polarix, lower_p_polarix, upper_p_polarix = CM.modelPerformance(acc_polarix, kappa_polarix, 
                                                                  lower_acc_polarix, upper_acc_polarix, 
                                                                  lower_kappa_polarix, upper_kappa_polarix)


performances_all.loc['phoenix'] = [acc_polarix, lower_acc_polarix, upper_acc_polarix,
                                  kappa_polarix, lower_kappa_polarix, upper_kappa_polarix,
                                  p_polarix, lower_p_polarix, upper_p_polarix]
```
