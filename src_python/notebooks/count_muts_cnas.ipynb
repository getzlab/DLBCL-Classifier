```python
muts_per_sample = maf['Tumor_Sample_Barcode'].value_counts().sort_values()
```

```python
def calc_cna_perc(sample):
    tmpdf = segs.loc[sample]
    total_cov = sum(tmpdf['End'] - tmpdf['Start'] + 1)

    del_cov = tmpdf.loc[tmpdf['Segment_Mean'] < -0.1]
    del_cov = sum(del_cov['End'] - del_cov['Start'] + 1)

    amp_cov = tmpdf.loc[tmpdf['Segment_Mean'] > 0.1]
    amp_cov = sum(amp_cov['End'] - amp_cov['Start'] + 1)
    
    del_frac = del_cov / total_cov
    amp_frac = amp_cov / total_cov
    
    return del_frac, amp_frac
```

```python
calc_cna_perc('DFCIDL001_DT')
```

```python
cna_fracs = {}
for s in set(segs.index):
    d_f, a_f = calc_cna_perc(s)
    cna_fracs[s] = [d_f, a_f]
```

```python
muts_per_sample = muts_per_sample.loc[samples]
muts_per_sample
```

```python
cna_df = pd.DataFrame.from_dict(cna_fracs, orient='index')
cna_df = cna_df.loc[samples]
cna_df.columns = ['del_frac', 'amp_frac']
cna_df['muts'] = muts_per_sample
cna_df
```
