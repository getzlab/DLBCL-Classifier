```python
import sys
sys.path.append("../")
import pandas as pd
import calculate_model_metrics
import numpy as np
import random
import format_data as fd
import classify_generic as cg
import calculate_model_metrics as CM
from matplotlib import pyplot as plt

random.seed(123)
np.random.seed(123)
```

```python
f1_cdx = pd.read_csv('../../data_tables/panel_gene_sets/f1_cdx_raw.txt', header=None)
f1_cdx = f1_cdx[0].str.replace('(', '').str.replace(')', '').str.split(' ')
all_f1_cdx = []
for row in f1_cdx:
    [all_f1_cdx.append(x.replace('-', '.')) for x in row]
    
pd.DataFrame(all_f1_cdx).to_csv('../../data_tables/panel_gene_sets/f1_cdx_fixed.txt', 
                                sep='\t', index=False, header=None)
```

```python
f1_heme = pd.read_csv('../../data_tables/panel_gene_sets/f1_heme_raw.txt', header=None)
f1_heme = f1_heme[0].str.replace('(', '').str.replace(')', '').str.split(' ')
all_f1_heme = []
for row in f1_heme:
    [all_f1_heme.append(x.replace('-', '.')) for x in row]
    
pd.DataFrame(all_f1_heme).to_csv('../../data_tables/panel_gene_sets/f1_heme_fixed.txt', 
                                sep='\t', index=False, header=None)
```

```python
all_msk_impact = list(pd.read_csv('../../data_tables/panel_gene_sets/msk_impact_raw.txt', 
                                  sep='\t', index_col=0, header=None).index)
all_lacy = list(pd.read_csv('../../data_tables/panel_gene_sets/lacy_raw.txt', 
                            sep='\t', index_col=0, header=None).index)

# some fixes
all_msk_impact = [x.replace('-', '.') for x in all_msk_impact]
all_lacy = [x.replace('-', '.') for x in all_lacy]

if 'MYD88' in all_f1_cdx:
    all_f1_cdx.append('MYD88.OTHER')
    all_f1_cdx.append('MYD88.L265P')
    
if 'MYD88' in all_f1_heme:
    all_f1_heme.append('MYD88.OTHER')
    all_f1_heme.append('MYD88.L265P')
    
if 'MYD88' in all_msk_impact:
    all_msk_impact.append('MYD88.OTHER')
    all_msk_impact.append('MYD88.L265P')
    
if 'MYD88' in all_lacy:
    all_lacy.append('MYD88.OTHER')
    all_lacy.append('MYD88.L265P')
```

```python
full_gsm = pd.read_csv('../../data_tables/gsm/DLBCL.699.163drivers.Sep_23_2022.tsv', sep='\t', index_col=1)
full_gsm = full_gsm.drop('gene', axis=1)
full_gsm
```

```python
means = full_gsm.mean(axis=1)
```

```python
full_gsm_f1_cdx = full_gsm.copy(deep=True)
full_gsm_f1_heme = full_gsm.copy(deep=True)
full_gsm_msk = full_gsm.copy(deep=True)
full_gsm_lacy = full_gsm.copy(deep=True)
```

```python
%cd ..
```

```python
# Zero out missing features from each panel
f1_cdx_zerod = []
for g in full_gsm_f1_cdx.index:
    if g.startswith('SV.') or g.endswith('.AMP') or g.endswith('.DEL'):
        continue
    if g not in all_f1_cdx:
        f1_cdx_zerod.append(g)
        full_gsm_f1_cdx.loc[g] = means.loc[g]
        
f1_heme_zerod = []
for g in full_gsm_f1_heme.index:
    if g.startswith('SV.') or g.endswith('.AMP') or g.endswith('.DEL'):
        continue
    if g not in all_f1_heme:
        f1_heme_zerod.append(g)
        full_gsm_f1_heme.loc[g] = means.loc[g]
        
msk_zerod = []
for g in full_gsm_msk.index:
    if g.startswith('SV.') or g.endswith('.AMP') or g.endswith('.DEL'):
        continue
    if g not in all_msk_impact:
        msk_zerod.append(g)
        full_gsm_msk.loc[g] = means.loc[g]

lacy_zerod = []
for g in full_gsm_lacy.index:
    if g.startswith('SV.') or g.endswith('.AMP') or g.endswith('.DEL'):
        continue
    if g not in all_lacy:
        lacy_zerod.append(g)
        full_gsm_lacy.loc[g] = means.loc[g]
```

```python
full_gsm_lacy
```

```python
labels = pd.read_csv('../data_tables/confidence_tables/baseline_probabilities.connectivity_based.sensitivity_power2.Sep_23_2022.tsv',
                    sep='\t', index_col=0)
labels['cluster'] = 'C' + labels['cluster'].astype(str)
labels = labels.loc[full_gsm_f1_cdx.columns]
labels
```

```python
# Make reduced gsms

reduced_gsm_f1_cdx = fd.construct_reduced_winning_version(full_gsm_f1_cdx)
classified_samples_f1_cdx = cg.classify_samples_winning_model(reduced_gsm_f1_cdx)
classified_samples_f1_cdx.index.name = 'sample'
classified_samples_f1_cdx['PredictedCluster'] = classified_samples_f1_cdx['PredictedCluster'].map({1.0: 'C1', 
                                                                                                   2.0: 'C2', 
                                                                                                   3.0: 'C3', 
                                                                                                   4.0: 'C4', 
                                                                                                   5.0: 'C5'})
classified_samples_f1_cdx['TrueCluster'] = labels['cluster']
classified_samples_f1_cdx = classified_samples_f1_cdx.sort_values(by='Confidence')
classified_samples_f1_cdx.to_csv('../evaluation_panel_sets/f1_cdx_predictions.tsv', sep='\t', index=True)
```

```python
# Make reduced gsms

reduced_gsm_f1_heme = fd.construct_reduced_winning_version(full_gsm_f1_heme)
classified_samples_f1_heme = cg.classify_samples_winning_model(reduced_gsm_f1_heme)
classified_samples_f1_heme.index.name = 'sample'
classified_samples_f1_heme['PredictedCluster'] = classified_samples_f1_heme['PredictedCluster'].map({1.0: 'C1', 
                                                                                                   2.0: 'C2', 
                                                                                                   3.0: 'C3', 
                                                                                                   4.0: 'C4', 
                                                                                                   5.0: 'C5'})
classified_samples_f1_heme['TrueCluster'] = labels['cluster']
classified_samples_f1_heme = classified_samples_f1_heme.sort_values(by='Confidence')
classified_samples_f1_heme.to_csv('../evaluation_panel_sets/f1_heme_predictions.tsv', sep='\t', index=True)
```

```python
# Make reduced gsms

reduced_gsm_msk = fd.construct_reduced_winning_version(full_gsm_msk)
classified_samples_msk = cg.classify_samples_winning_model(reduced_gsm_msk)
classified_samples_msk.index.name = 'sample'
classified_samples_msk['PredictedCluster'] = classified_samples_msk['PredictedCluster'].map({1.0: 'C1', 
                                                                                                   2.0: 'C2', 
                                                                                                   3.0: 'C3', 
                                                                                                   4.0: 'C4', 
                                                                                                   5.0: 'C5'})
classified_samples_msk['TrueCluster'] = labels['cluster']
classified_samples_msk = classified_samples_msk.sort_values(by='Confidence')
classified_samples_msk.to_csv('../evaluation_panel_sets/msk_predictions.tsv', sep='\t', index=True)
```

```python
# Make reduced gsms

reduced_gsm_lacy = fd.construct_reduced_winning_version(full_gsm_lacy)
classified_samples_lacy = cg.classify_samples_winning_model(reduced_gsm_lacy)
classified_samples_lacy.index.name = 'sample'
classified_samples_lacy['PredictedCluster'] = classified_samples_lacy['PredictedCluster'].map({1.0: 'C1', 
                                                                                               2.0: 'C2', 
                                                                                               3.0: 'C3', 
                                                                                               4.0: 'C4', 
                                                                                               5.0: 'C5'})
classified_samples_lacy['TrueCluster'] = labels['cluster']
classified_samples_lacy = classified_samples_lacy.sort_values(by='Confidence')
classified_samples_lacy.to_csv('../evaluation_panel_sets/lacy_predictions.tsv', sep='\t', index=True)
```

```python
performances_all = pd.DataFrame(columns=['accuracy', 'lowerAcc', 'upperAcc', 
                                         'kappa', 'lowerKappa', 'upperKappa',
                                         'performance', 'lowerPerformance', 'upperPerformance'])
```

```python
# f1 cdx results

lrx_f1_cdx = np.array(classified_samples_f1_cdx['Confidence'])
lry_f1_cdx = np.array(np.equal(classified_samples_f1_cdx['PredictedCluster'], classified_samples_f1_cdx['TrueCluster']).astype(int))
CM.modelKappa(lrx_f1_cdx, lry_f1_cdx,
              num_iter=1,
              computeplots=True,
              savefilename='../plots/xyresiduals/f1_cdx')

acc_f1_cdx, lower_acc_f1_cdx, upper_acc_f1_cdx = CM.modelAccuracy(classified_samples_f1_cdx['PredictedCluster'], 
                                                                  classified_samples_f1_cdx['TrueCluster'])
kappa_f1_cdx, lower_kappa_f1_cdx, upper_kappa_f1_cdx = CM.modelKappa(lrx_f1_cdx, lry_f1_cdx)
p_f1_cdx, lower_p_f1_cdx, upper_p_f1_cdx = CM.modelPerformance(acc_f1_cdx, kappa_f1_cdx, lower_acc_f1_cdx, upper_acc_f1_cdx, lower_kappa_f1_cdx, upper_kappa_f1_cdx)


performances_all.loc['f1_cdx'] = [acc_f1_cdx, lower_acc_f1_cdx, upper_acc_f1_cdx,
                                  kappa_f1_cdx, lower_kappa_f1_cdx, upper_kappa_f1_cdx,
                                  p_f1_cdx, lower_p_f1_cdx, upper_p_f1_cdx]
```

```python
# f1 heme results

classified_samples_f1_heme = classified_samples_f1_heme.sort_values(by='Confidence')

lrx_f1_heme = np.array(classified_samples_f1_heme['Confidence'])
lry_f1_heme = np.array(np.equal(classified_samples_f1_heme['PredictedCluster'], classified_samples_f1_heme['TrueCluster']).astype(int))
CM.modelKappa(lrx_f1_heme, lry_f1_heme,
              num_iter=1,
              computeplots=True,
              savefilename='../plots/xyresiduals/f1_heme')

acc_f1_heme, lower_acc_f1_heme, upper_acc_f1_heme = CM.modelAccuracy(classified_samples_f1_heme['PredictedCluster'], 
                                                                     classified_samples_f1_heme['TrueCluster'])
kappa_f1_heme, lower_kappa_f1_heme, upper_kappa_f1_heme = CM.modelKappa(lrx_f1_heme, lry_f1_heme)
p_f1_heme, lower_p_f1_heme, upper_p_f1_heme = CM.modelPerformance(acc_f1_heme, kappa_f1_heme, lower_acc_f1_heme, 
                                                                  upper_acc_f1_heme, lower_kappa_f1_heme, upper_kappa_f1_heme)


performances_all.loc['f1_heme'] = [acc_f1_heme, lower_acc_f1_heme, upper_acc_f1_heme,
                                  kappa_f1_heme, lower_kappa_f1_heme, upper_kappa_f1_heme,
                                  p_f1_heme, lower_p_f1_heme, upper_p_f1_heme]
```
