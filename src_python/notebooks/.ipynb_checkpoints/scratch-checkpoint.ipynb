```python
import pandas as pd
```

```python
tcga_svs = pd.read_csv('../../data_tables/gsm/old_matrices/DLBCL_Staudt_Shipp_CL.for_classifier_training.classifier_subset.fix_sv.fix_ploidy.CLUSTERING.17-Aug-2022.txt',
                      sep='\t', index_col=0)

sample_set = pd.read_csv('../../data_tables/sample_sets/ShippStaudtSets.purity0.2.txt', sep='\t', index_col=0)
tcga_samples = sample_set.loc[sample_set['cohort'] == 'TCGA']
```

```python
tcga_svs
```

```python
tcga_svs = tcga_svs.loc[['SV.BCL2', 'SV.BCL6', 'SV.MYC'], tcga_svs.columns.isin(tcga_samples.index)]
cp_df = tcga_svs.copy(deep=True)
cp_df.index = cp_df.index + '.CCF' 
tcga_svs = pd.concat([tcga_svs, cp_df])
tcga_svs.loc[['SV.BCL2.CCF', 'SV.BCL6.CCF', 'SV.MYC.CCF']] = tcga_svs.loc[['SV.BCL2.CCF', 'SV.BCL6.CCF', 'SV.MYC.CCF']].astype(bool).astype(int)
print(tcga_svs.columns)
tcga_svs.to_csv('../../data_tables/gsm/tcga_svs.tsv', sep='\t', index=True)
```
