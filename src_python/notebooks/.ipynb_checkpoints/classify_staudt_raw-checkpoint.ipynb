```python
our_gsm = pd.read_csv('../data_tables/gsm/DLBCL.699.163drivers.Sep_23_2022.tsv', sep='\t', index_col=1)
our_gsm
```

```python
sample_set = pd.read_csv('../data_tables/sample_sets/ShippStaudtSets.purity0.2.txt', sep='\t', index_col=0)
staudt_samples = sample_set[sample_set['cohort'] == 'Staudt']
tcga_samples = sample_set[sample_set['cohort'] == 'TCGA']
shipp_samples = sample_set[sample_set['cohort'] == 'Shipp']
```

```python
staudt_samples
```

```python
mutation_rows = [x for x in our_gsm.index if '.AMP' not in x and '.DEL' not in x and 'SV.' not in x]
len(mutation_rows)
```

```python
staudt_gsm = our_gsm.loc[:, our_gsm.columns.isin(staudt_samples.index)].copy(deep=True)
staudt_gsm.loc[mutation_rows, :] = 0
staudt_gsm
```

```python
tcga_gsm = our_gsm.loc[:, our_gsm.columns.isin(tcga_samples.index)].copy(deep=True)
tcga_gsm.loc[mutation_rows, :] = 0
tcga_gsm
```

```python
for idx, row in staudt_maf.iterrows():
    sample = row['SUBJECT_NAME']
    gene = row['GENE SYMBOL'].replace('-', '.')
    if sample not in staudt_gsm.columns or gene not in staudt_gsm.index:
        continue
    
    if row['GENE SYMBOL'] == 'MYD88':
        if row['AA CHANGE'] == 'L265P':
            staudt_gsm.loc['MYD88.L265P', sample] = 2
        else:
            staudt_gsm.loc['MYD88.OTHER', sample] = 2
    
    staudt_gsm.loc[gene, sample] = 2
```

```python
for idx, row in tcga_maf.iterrows():
    sample = row['gsm_name']
    gene = row['GENE SYMBOL'].replace('-', '.')
    if sample not in tcga_gsm.columns or gene not in tcga_gsm.index:
        continue
    
    if row['GENE SYMBOL'] == 'MYD88':
        if row['AA CHANGE'] == 'L265P':
            tcga_gsm.loc['MYD88.L265P', sample] = 2
        else:
            tcga_gsm.loc['MYD88.OTHER', sample] = 2
    
    tcga_gsm.loc[gene, sample] = 2
```

```python
staudt_ours = our_gsm.loc[:, our_gsm.columns.isin(staudt_samples.index)].copy(deep=True)
staudt_ours
```

```python
labels = pd.read_csv('../data_tables/confidence_tables/baseline_probabilities.connectivity_based.sensitivity_power2.Sep_23_2022.tsv',
                    sep='\t', index_col=0)
labels['cluster'] = 'C' + labels['cluster'].astype(str)
labels = labels.loc[staudt_gsm.columns]
labels
```

```python
test_set = list(pd.read_csv('../data_tables/train_test_sets/TestingSet_149Subset_May2021.txt', 
                            sep='\t', index_col=0, header=None).index)
```

```python
# classify staudt_ours and staudt_raw

# Make reduced gsms

staudt_ours_test = staudt_ours.loc[:, staudt_ours.columns.isin(test_set)]

reduced_gsm_staudt_ours = fd.construct_reduced_winning_version(staudt_ours_test)
classified_samples_staudt_ours = cg.classify_samples_winning_model(reduced_gsm_staudt_ours)
classified_samples_staudt_ours.index.name = 'sample'
classified_samples_staudt_ours['PredictedCluster'] = classified_samples_staudt_ours['PredictedCluster'].map({
                                                                                                     1.0: 'C1', 
                                                                                                     2.0: 'C2', 
                                                                                                     3.0: 'C3', 
                                                                                                     4.0: 'C4', 
                                                                                                     5.0: 'C5'})
classified_samples_staudt_ours['TrueCluster'] = labels.loc[classified_samples_staudt_ours.index, 'cluster']
classified_samples_staudt_ours = classified_samples_staudt_ours.sort_values(by='Confidence')
classified_samples_staudt_ours.to_csv('../evaluation_panel_sets/staudt_ours_predictions.tsv', sep='\t', index=True)
```

```python
staudt_raw_test = staudt_gsm.loc[:, staudt_gsm.columns.isin(test_set)]

reduced_gsm_staudt_raw = fd.construct_reduced_winning_version(staudt_raw_test)
classified_samples_staudt_raw = cg.classify_samples_winning_model(reduced_gsm_staudt_raw)
classified_samples_staudt_raw.index.name = 'sample'
classified_samples_staudt_raw['PredictedCluster'] = classified_samples_staudt_raw['PredictedCluster'].map({
                                                                                                     1.0: 'C1', 
                                                                                                     2.0: 'C2', 
                                                                                                     3.0: 'C3', 
                                                                                                     4.0: 'C4', 
                                                                                                     5.0: 'C5'})
classified_samples_staudt_raw['TrueCluster'] = labels.loc[classified_samples_staudt_raw.index, 'cluster']
classified_samples_staudt_raw = classified_samples_staudt_raw.sort_values(by='Confidence')
classified_samples_staudt_raw.to_csv('../evaluation_panel_sets/staudt_raw_predictions.tsv', sep='\t', index=True)
```

```python
performances_all = pd.DataFrame(columns=['accuracy', 'lowerAcc', 'upperAcc', 
                                         'kappa', 'lowerKappa', 'upperKappa',
                                         'performance', 'lowerPerformance', 'upperPerformance'])
```

```python
# staudt our calls results

lrx_staudt_ours = np.array(classified_samples_staudt_ours['Confidence'])
lry_staudt_ours = np.array(np.equal(classified_samples_staudt_ours['PredictedCluster'], 
                                classified_samples_staudt_ours['TrueCluster']).astype(int))
CM.modelKappa(lrx_staudt_ours, lry_staudt_ours,
              num_iter=1,
              computeplots=True,
              savefilename='../plots/xyresiduals/staudt_ours')

acc_staudt_ours, lower_acc_staudt_ours, upper_acc_staudt_ours = CM.modelAccuracy(
                                                                    classified_samples_staudt_ours['PredictedCluster'], 
                                                                    classified_samples_staudt_ours['TrueCluster'])

kappa_staudt_ours, lower_kappa_staudt_ours, upper_kappa_staudt_ours = CM.modelKappa(lrx_staudt_ours, 
                                                                                    lry_staudt_ours)

p_staudt_ours, lower_p_staudt_ours, upper_p_staudt_ours = CM.modelPerformance(acc_staudt_ours, kappa_staudt_ours, 
                                                                  lower_acc_staudt_ours, upper_acc_staudt_ours, 
                                                                  lower_kappa_staudt_ours, upper_kappa_staudt_ours)


performances_all.loc['staudt_ours'] = [acc_staudt_ours, lower_acc_staudt_ours, upper_acc_staudt_ours,
                                  kappa_staudt_ours, lower_kappa_staudt_ours, upper_kappa_staudt_ours,
                                  p_staudt_ours, lower_p_staudt_ours, upper_p_staudt_ours]
```

```python
# staudt raw calls results

lrx_staudt_raw = np.array(classified_samples_staudt_raw['Confidence'])
lry_staudt_raw = np.array(np.equal(classified_samples_staudt_raw['PredictedCluster'], 
                                classified_samples_staudt_raw['TrueCluster']).astype(int))
CM.modelKappa(lrx_staudt_raw, lry_staudt_raw,
              num_iter=1,
              computeplots=True,
              savefilename='../plots/xyresiduals/staudt_raw')

acc_staudt_raw, lower_acc_staudt_raw, upper_acc_staudt_raw = CM.modelAccuracy(
                                                                    classified_samples_staudt_raw['PredictedCluster'], 
                                                                    classified_samples_staudt_raw['TrueCluster'])

kappa_staudt_raw, lower_kappa_staudt_raw, upper_kappa_staudt_raw = CM.modelKappa(lrx_staudt_raw, 
                                                                                    lry_staudt_raw)

p_staudt_raw, lower_p_staudt_raw, upper_p_staudt_raw = CM.modelPerformance(acc_staudt_raw, kappa_staudt_raw, 
                                                                  lower_acc_staudt_raw, upper_acc_staudt_raw, 
                                                                  lower_kappa_staudt_raw, upper_kappa_staudt_raw)


performances_all.loc['staudt_raw'] = [acc_staudt_raw, lower_acc_staudt_raw, upper_acc_staudt_raw,
                                  kappa_staudt_raw, lower_kappa_staudt_raw, upper_kappa_staudt_raw,
                                  p_staudt_raw, lower_p_staudt_raw, upper_p_staudt_raw]
```
