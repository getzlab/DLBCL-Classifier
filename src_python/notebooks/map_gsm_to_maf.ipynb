```python
maf = pd.read_csv('../../data_tables/maf_files/DLBCL_combined_700.hg38B.trimmed.maf', sep='\t')
maf.head()
```

```python
sig_genes = pd.read_csv('../../data_tables/qval_dfs/fisher_exact_5x2_17-Aug-2022.combined.tsv', 
                        sep='\t', index_col=0)
sig_genes = sig_genes.loc[sig_genes['q'] <= 0.10].index

sig_genes
```

```python
amps = [x for x in sig_genes if '.AMP' in x]
dels = [x for x in sig_genes if '.DEL' in x]
svs = [x for x in sig_genes if 'SV.' in x]

mutations = [x for x in sig_genes if (x not in amps) and (x not in dels) and (x not in svs)]
```

```python
maf['Variant_Classification'].value_counts()
```

```python
# if maf changes (?) make sure you re-do these. Whitelisting coding is a bit safer than blacklisting noncoding,
# since UNDETERMINED & other things may show up that we don't want.
coding = ['Missense_Mutation', 'Splice_Site', 'Nonsense_Mutation', 'Frame_Shift_Del', 'Frame_Shift_Ins',
          'In_Frame_Del', 'In_Frame_Ins', 'Start_Codon_SNP', 'Nonstop_Mutation', 'De_novo_Start_InFrame',
          'De_novo_Start_OutOfFrame', 'Start_Codon_Ins', 'Stop_Codon_Del', 'Start_Codon_Del', 'Stop_Codon_Ins']
```

```python
gsm = pd.read_csv('../../data_tables/gsm/DLBCL_Staudt_Shipp_CL.for_classifier_training.classifier_subset.fix_sv.fix_ploidy.17-Aug-2022.txt',
                 sep='\t', index_col=0)
gsm = gsm.loc[sig_genes].astype(float).astype(int)
gsm.head()
```

```python
maf_coding = maf.loc[maf['Variant_Classification'].isin(coding + ['Silent'])]
maf_coding
```

```python
gsm_ac = pd.DataFrame(np.nan, index=mutations, columns=gsm.columns)
gsm_ac
```

```python
maf_coding.loc[:, 'Hugo_Symbol'] = maf_coding.loc[:, 'Hugo_Symbol'].str.replace('-', '.')
```

```python
myd88_l265p = {'p.L265P', 'p.*160R', 'p.L273P'}
for col in gsm.columns:
    sample_col = gsm[col]
    sample = sample_col.name
    events = sample_col[sample_col != 0]
    maf_s = maf_coding.loc[maf_coding['Tumor_Sample_Barcode'] == sample]
    
    for e in events.index:
        if e not in mutations:
            continue
            
        val = events[e]
        
        if e == 'MYD88':
            continue
        elif e == 'MYD88.L265P':
            maf_e = maf_s.loc[(maf_s['Hugo_Symbol'] == 'MYD88') & 
                              (maf_s['Protein_Change'].isin(myd88_l265p))]
        elif e == 'MYD88.OTHER':
            maf_e = maf_s.loc[(maf_s['Hugo_Symbol'] == 'MYD88') & 
                              ~(maf_s['Protein_Change'].isin(myd88_l265p))]
        else:
            maf_e = maf_s.loc[maf_s['Hugo_Symbol'] == e]
        
        if val == 1:
            maf_e = maf_e.loc[maf_e['Variant_Classification'] == 'Silent']
        else:
            maf_e = maf_e.loc[maf_e['Variant_Classification'] != 'Silent']
            
        max_talt = maf_e['t_alt_count'].max()
        
        gsm_ac.loc[e, sample] = max_talt
```

```python
gsm_ac
```

```python
gsm_sv_reads = pd.DataFrame(np.nan, index=svs, columns=gsm_ac.columns)
gsm_sv_reads
```

```python
sv_df['VCF_TALT']
```

```python
for col in gsm.columns:
    sample_col = gsm[col]
    sample = sample_col.name
    events = sample_col[['SV.BCL2', 'SV.BCL6', 'SV.MYC']]
    if not sum(events):
        continue
        
    sv_s = sv_df.loc[sv_df['individual'] == sample]
    
    if not len(sv_s.index):
        continue
        
    for e in events.index:
        file_event = e.replace('SV.', '')
        val = events[e]
        
        if not val:
            continue
        
        sv_e = sv_s.loc[(sv_s['gene1'] == file_event) | (sv_s['gene2'] == file_event)]
        max_treads = sv_e['VCF_TALT'].max()
        gsm_sv_reads.loc[e, sample] = max_treads
        
        if np.isnan(max_treads):
            print(sample, e)
    
```

```python
gsm_mut_svs = pd.concat([gsm_ac, gsm_sv_reads])
gsm_mut_svs
```

```python
gsm_mut_svs.to_csv('../../data_tables/gsm/gsm_17-Aug-2022_altcounts.tsv', sep='\t')
```
